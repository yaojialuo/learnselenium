```python
quote=pd.read_csv(r"E:\orderflow\MarketQuote\2021-07-05\002335.csv",index_col=0)

quote['created_at']=quote.created_at.apply(lambda x:x[-15:-6])
quote['low']=quote['low'].apply(lambda x:float('%0.2f'%x))
quote['high']=quote['high'].apply(lambda x:float('%0.2f'%x))
quote['price']=quote['price'].apply(lambda x:float('%0.2f'%x))
for i in range(5):
    quote[f"bp{i+1}"]=quote.quotes.apply(lambda x:float('%0.2f'%eval(x)[i]['bid_p']) )
    quote[f"bv{i+1}"]=quote.quotes.apply(lambda x:eval(x)[i]['bid_v'] )
    quote[f"ap{i+1}"]=quote.quotes.apply(lambda x:float('%0.2f'%eval(x)[i]['ask_p']) )
    quote[f"av{i+1}"]=quote.quotes.apply(lambda x:eval(x)[i]['ask_v'] )
quote
```

```python
quote.drop(['symbol','quotes'],axis=1,inplace=True)
```
